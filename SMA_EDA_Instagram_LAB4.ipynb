```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Instagram_Analytics.csv")
data.head()
```

```python
data.info()
data.describe()
```

```python
data.columns
```

```python
data.groupby("media_type")["likes"].mean().plot(kind="bar")
plt.title("Average Likes by Post Type")
plt.xlabel("Post Type")
plt.ylabel("Likes")
plt.show()
```
